**1.2 Analysis Design: Theory and Hypotheses**

---

The primary goal of this section is to understand the factors that influence cab drivers' revenue. The following research questions and hypotheses will guide our analysis:


**Research Questions**

This study aims to answer following question by implementing different stastistical methods:


*   What are the primary factors that influence the revenue generated by taxi rides?
*   How do different payment methods affect tipping behaviors?
*   Is there a correlation between trip distance and fare amounts?
*   How do congestion surcharges impact overall fares?
*   Are trip durations and fare amounts affected by the time of day?


**Hypotheses**


This mini-project will focus on several key hypotheses:


**H1: Trip Distance and Fare Amount**
- Longer trip distances will result in higher fares, as taxi fare calculations typically include distance-based pricing.

**H2: Passenger Count and Fare/Tipping Behavior**
- Trips with more passengers may have higher fares due to potential group sharing or surcharge rules. Additionally, groups might tip differently than solo passengers.

**H3: Payment Type and Tip Amount**
- Credit card payments might show higher tips due to convenience features like pre-set tipping percentages on card terminals, while cash payments may show more variability in tip amounts.

**H4: Time of Day and Trip Duration/Fare**
- Nighttime rides may have higher fares due to surcharges and potentially lower traffic congestion.  
- Peak traffic times during the day could increase trip durations and fares due to congestion charges.

**H5: Congestion Surcharges and Total Fare Amount**
- Trips taken during high-congestion times or areas will have higher total fares due to additional charges, regardless of distance traveled.

**1.3 Sampling, Validity and Reliability of the Planned Analysis**

---

The dataset used for this analysis comprises records of cab rides collected over a specific period. This section covers the sampling methods and the analysis's validity to ensure our findings' reliability.

**Sampling**

The NYC Taxi Fare Dataset from Kaggle was chosen for this project. It provides detailed records of taxi trips, making it suitable for answering my research questions. However, since the data is a convenience sample (collected from a specific source rather than randomly), it may include biases. For instance, there might be more trips from certain areas or times of day, which could affect the results.

**Validity**

To ensure the analysis is valid, I planned to carefully handle missing or incorrect data, such as negative fares or distances, and exclude unrealistic values, like very long trips with low fares. Additionally, I focused on selecting variables that are directly relevant to my hypotheses, like trip distance, fare amount, and passenger count.

**Reliability**

For reliability, I aimed to use consistent and appropriate statistical methods, starting with assumption checks before choosing the right tests. To further enhance the reliability of my results, techniques like bootstrapping were included to validate the findings. This approach ensures that the analysis is both trustworthy and focused on answering the research questions.

By addressing these points, I aim to conduct a thorough and reliable analysis that provides meaningful insights into what affects cab drivers' revenue. This careful approach ensures that my findings are valid, reliable, and useful in real-world scenarios.
