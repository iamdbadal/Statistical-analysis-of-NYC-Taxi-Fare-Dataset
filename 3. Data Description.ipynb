# 3. Descriptive Analysis

**Data Preview**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('preprocessed_data.csv')
df.info()
df.describe()
```

The dataset contains 99,645 taxi trip records and includes a variety of taxi trip-related variables, such as VendorID, passenger_count, trip_distance, fare_amount, total_amount, and several surcharges. Out of these, I have categorised them into critical and non-critical fields. These are listed below:

**Critical Fields include:** These columns are responsible for answering research questions and hypothesis.

`trip_distance`,`passenger_count`,`fare_amount`,`total_amount`,`tip_amount`,`tpep_pickup_datetime`,`tpep_dropoff_datetime`,`payment_type`, `congestion_surcharge`, `trip_duration` and `hour_of_day`.

**Non-Critical Fields include:**

`VendorID`,`RatecodeID`,`store_and_fwd_flag`,`PULocationID`,`DOLocationID`,`extra`,`mta_tax`,`tolls_amount` and `improvement_surcharge`.

A preliminary analysis of these key variables provides the following insights:

**Trip Distance:**

The average trip distance is 3.14 miles, with a standard deviation of 4.23 miles, indicating substantial variability in trip lengths. The range spans from 0.2 miles to 259.22 miles, with the majority of trips being relatively short but some representing long-distance or intercity travel.

**Passenger Count:**

The mean passenger count per trip is approximately 1.58, with a standard deviation of 1.19, indicating most trips involve one or two passengers. The range spans from 0 to 9 passengers, with larger group trips being less common.

**Fare Amount:**

The average fare amount is 13.07, with a standard deviation of 13.46, reflecting a wide range of fares. The fare amount ranges from 0 to 1238.00, suggesting that while many fares are relatively low, there are significant variations due to longer trips or other factors.

**Total Amount:**

The total amount averages 18.91, with a standard deviation of 16.09, indicating significant variation in the total fare due to additional surcharges and tips. The range for the total amount spans from 0 to $1242.30, showing that total fares can vary greatly depending on the trip length and additional charges.

**Tip Amount:**

The average tip amount is 2.09, with a standard deviation of 2.80, indicating variability in tipping behavior. Tip amounts range from 0 to $164.70, with higher tips typically associated with longer or more expensive trips.

**Trip Duration (trip_duration):**

The mean trip duration is 15.44 minutes, with a standard deviation of 65.36 minutes, showing a wide range of durations. The minimum duration is 0 minutes, indicating possible errors or incomplete data, while the maximum duration is 2458.4 minutes, which is likely an outlier caused by delays or errors in data entry.

**Pick-up Hour (hour_of_day):**

The mean pick-up hour is 13.22 hours, with the range spanning from 0 to 23 hours. This suggests that trips occur throughout the day, with a peak in the afternoon (around 2 PM), and reflects typical daily taxi usage patterns.

**Payment Type:**

The mean payment type value is 1.32, likely indicating a mix of payment methods (e.g., cash, credit card). The distribution of payment types helps identify trends in how passengers pay for their trips, which may also correlate with tipping behavior and trip length.

**Congestion Surcharge:**

The mean congestion surcharge is $2.27, with a standard deviation of 0.73, indicating moderate variability in surcharge application. The surcharge ranges from -2.50 to 2.75, suggesting that it is typically applied during peak traffic times, although the negative values should be investigated for potential data errors.

**Key Observations:**

* **Trip distances** and **trip durations** show significant variability, with most trips being short but some representing long-distance travel. The presence of extreme values and outliers may require further investigation.
* **Fare amounts** and **total amounts** also exhibit wide variability, with significant differences between short and long trips, as well as the impact of surcharges and tips.
* **Tips** show a broad range, likely influenced by trip characteristics such as duration, distance, and fare amount.
* **Congestion surcharges** reflect additional charges applied during peak traffic conditions, contributing to higher total fares.

### Visualization

Visualization of the critical fileld explain the patterns in the critical field of the dataset. Which is helpful for the statistical analysis.

```python
# Setting plot style
sns.set(style="whitegrid")

# Critical fields
critical_fields = [
    'trip_distance', 'passenger_count', 'fare_amount', 'total_amount', 'tip_amount',
    'tpep_pickup_datetime', 'tpep_dropoff_datetime', 'payment_type', 'congestion_surcharge',
    'trip_duration', 'hour_of_day'
]

# Filter dataframe to include only critical fields
df_critical = df[critical_fields]

# 1. Pairplot for numeric variables to see relationships
numeric_cols = df_critical.select_dtypes(include=['float64', 'int64']).columns
if len(numeric_cols) > 1:
    sns.pairplot(df_critical[numeric_cols], diag_kind='kde')
    plt.suptitle('Pairplot of Numeric Variables', y=1.02)
    plt.show()

# 2. Tip Amount by Payment Type
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_critical['payment_type'], y=df_critical['tip_amount'])
plt.title('Tip Amount by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Tip Amount')
plt.xticks(rotation=45)
plt.show()

# 3. Trip Distance vs Fare Amount
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_critical['trip_distance'], y=df_critical['fare_amount'])
plt.title('Trip Distance vs Fare Amount')
plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.show()

# 4. Total Amount by Congestion Surcharge
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_critical['congestion_surcharge'], y=df_critical['total_amount'])
plt.title('Total Amount by Congestion Surcharge')
plt.xlabel('Congestion Surcharge')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()

# 5. Trip Duration by Hour of Day
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_critical['hour_of_day'], y=df_critical['trip_duration'])
plt.title('Trip Duration by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Trip Duration')
plt.xticks(rotation=45)
plt.show()

# Fare Amount by Hour of Day
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_critical['hour_of_day'], y=df_critical['fare_amount'])
plt.title('Fare Amount by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Fare Amount')
plt.xticks(rotation=45)
plt.show()
```
