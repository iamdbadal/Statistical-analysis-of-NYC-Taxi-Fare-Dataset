# 7. Conclusion

This statistical analysis explored and analyzed the relationships among several variables, such as trip distance, fare amount, passenger count, tip amount, and congestion surcharge, with the help of advanced statistical methods including bootstraping and Bayesian approaches. By addressing the research questions and hypotheses through thorough data analysis, we derive the following conclusions based on the findings:

**Research Question 1: What is the relationship between trip distance and fare amount?**

**Findings:**

Both the Pearson correlation analysis and Bayesian correlation confirmed a strong, positive relationship between trip distance and fare amount. The Pearson correlation coefficient was approximately 0.83 with a 95% confidence interval of [0.7707, 0.8858]. Bayesian analysis showed that the posterior mean of the correlation coefficient was approximately 1, with a very narrow 95% HDI around this value.

**Conclusion:**

Both frequentist and Bayesian results independently confirm a very strong positive correlation between trip distance and fare amount. This is intuitive and expected because fare is directly tied to trip length in most fare systems.


**Research Question 2: Does passenger count significantly impact the tip amount?**

**Findings:**

**Bayesian ANOVA:**

* Group means showed significant variations in tip amounts, with the group representing higher passenger counts (e.g., "group_means 7") showing a markedly higher tip mean (e.g., mean ~9.2).
  
* The 95% HDI intervals for the group means did not overlap substantially, confirming significant differences between groups.

**Conclusion:**

Bayesian ANOVA results provide strong evidence for the hypothesis that passenger count impacts tip amount. The higher tipping behavior for larger passenger groups aligns with plausible social dynamics (e.g., group sharing or generosity).

**Research Question 3: What is the relationship between congestion surcharge and total fare amount?**

**Findings:**

The frequentist regression model found a negative coefficient for congestion surcharge (-0.144), with a 95% confidence interval of [-0.219, -0.068], indicating a slight decrease in total fare when congestion surcharge is included. Bayesian regression analysis corroborated this finding, with a 95% HDI for the coefficient ranging from [-0.3629, 0.0503]. However, the Bayesian interval was wider, reflecting uncertainty about this relationship.

**Conclusion:**

The results partially support the hypothesis that congestion surcharge contributes to variations in the total fare. However, the negative relationship is counterintuitive and might indicate that other confounding factors (e.g., discounts, shorter trips) are influencing the results.

**Research Question 4: How well do trip distance, passenger count, and congestion surcharge predict total fare?**

**Findings:**

The multivariable regression model explained 71.5% of the variation in total fare (R² = 0.715). Among the predictors:

* Trip Distance: Showed a strong positive relationship with fare, with a coefficient of 3.1835 and narrow confidence intervals ([3.0943, 3.2496] in the bootstrap analysis).
  
* Passenger Count: The relationship was weak and statistically insignificant (p-value > 0.05), with confidence intervals covering zero in both frequentist and Bayesian analyses.
  
* Congestion Surcharge: The coefficient was slightly negative, with wide confidence intervals, indicating weak predictive power.
  
**Conclusion:**

Trip distance is the strongest and most reliable predictor of total fare. Passenger count and congestion surcharge have limited or no predictive power in this context. The hypothesis that these variables collectively explain fare is partially supported, as trip distance overwhelmingly dominates the predictive model.

**Research Question 5: What are the differences in analysis between frequentist and Bayesian approaches?**

**Findings:**

Both frequentist and Bayesian methods yielded similar conclusions for key relationships. However:

* Bayesian methods provided richer insights, such as posterior distributions and HDI intervals, which capture the uncertainty around parameter estimates more intuitively.
  
* Frequentist confidence intervals were narrower in some cases, possibly due to assumptions about data normality and fixed parameters.
  
**Conclusion:**

Bayesian methods complement frequentist approaches by allowing for a probabilistic interpretation of results. These methods were particularly helpful in understanding the uncertainty in regression coefficients and group mean comparisons.

Yes, our analysis clearly shows that the fare increases with the distance traveled. This fits our expectation, as fares are naturally higher for longer trips.

**2. Does the number of passengers influence the tip amount?**

Yes, the number of passengers does impact tips. Larger groups tend to leave higher tips, which aligns with the idea that people in groups may feel more generous or share tips among themselves.

**3. Does the congestion surcharge affect the total fare?**

The results are less clear here. While we expected the surcharge to increase total fares, the relationship appears weak and even slightly negative. This suggests other factors might be influencing this outcome, like shorter trips often being in congested areas.

**4. Can we use trip distance, passenger count, and congestion surcharge to predict total fares?**

Partially. Trip distance is the strongest factor in predicting fares, but passenger count and congestion surcharge contribute very little. Most of the fare variation can be explained by how far the trip is.

**5. Are Bayesian methods more insightful than traditional methods?**

Both approaches gave similar conclusions, but Bayesian methods provided a clearer understanding of uncertainty, helping us better interpret relationships like the congestion surcharge’s effect.
